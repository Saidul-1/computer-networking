Consider the continuous-time analog signal x(t)=3cos(100πt). Sample the analog
signal at 200 Hz and 75 Hz. Show the discrete-time signal after sampling. ⟹
realization

```python
import numpy as np
import matplotlib.pyplot as plt

t_max = 0.1
t = np.linspace(0, t_max, 1000)
cont_signal = 3*np.cos(100*np.pi*t)

f1 = 200
f2 = 75

n1 = np.arange(0, t_max, 1/f1)
x1 = 3*np.cos(2*50*np.pi*n1)

n2 = np.arange(0, t_max, 1/f2)
x2 = 3*np.cos(2*50*np.pi*n2)

plt.subplot(3, 1, 1)
plt.plot(t, cont_signal)
plt.title("Original Continuous-time signal")

plt.subplot(3, 1, 2)
plt.stem(n1, x1)
plt.title("Sampled at 200Hz")

plt.subplot(3, 1, 3)
plt.plot(t, cont_signal)
plt.stem(n2, x2)
plt.title("Sampled at 75Hz")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Analog signal
t_cont = np.linspace(0, 0.1, 1000)
# t_count = np.arange(0, 0.1, 0.001)
x_cont = 3 * np.cos(100 * np.pi * t_cont)

# Sampling at 200 Hz
fs1 = 200
t_samples1 = np.arange(0, 0.1, 1/fs1)
x_samples1 = 3 * np.cos(100 * np.pi * t_samples1)

# Sampling at 75 Hz
fs2 = 75
t_samples2 = np.arange(0, 0.1, 1/fs2)
x_samples2 = 3 * np.cos(100 * np.pi * t_samples2)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t_cont, x_cont, label='Analog Signal')
plt.stem(t_samples1, x_samples1, linefmt='r-', markerfmt='ro', basefmt=' ', label='Samples (200 Hz)')
plt.title('Sampling at 200 Hz (Above Nyquist Rate)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t_cont, x_cont, label='Analog Signal')
plt.stem(t_samples2, x_samples2, linefmt='r-', markerfmt='ro', basefmt=' ', label='Samples (75 Hz)')
plt.title('Sampling at 75 Hz (Below Nyquist Rate - Aliasing)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()

# Calculate aliased frequency
aliased_freq = np.abs(fs2 - 50)  # 75 - 50 = 25 Hz
print(f"Aliased frequency when sampling at 75 Hz: {aliased_freq} Hz")
```
