Generating elementary signals like Unit Step, Ramp, Exponential, Sine, and
Cosine sequences. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Unit Step Sequence
The Unit Step Sequence is a discrete-time signal defined as:

[ u(n) = 1 if n >=0 else 0 ]

It is commonly used in signal processing and control systems to represent a signal that "turns on" at ( n = 0 ).



```python
n = np.arange(-10, 11, 1) #[-10, 11) with step size 1
u_n = np.where(n >= 0, 1, 0) # 1 if n >= 0, else 0

plt.stem(n, u_n, label="Unit Step Seq")
plt.xlabel("n")
plt.ylabel("u_n")
plt.legend()
plt.show()
```

Ramp Sequence (r[n] or r(n)):

Definition: A discrete-time signal that is zero for n<0 and increases linearly with n for n≥0.

```python
n = np.arange(-10, 11, 1)
ramp = np.where(n < 0, 0, n)

plt.stem(n, ramp, label="Ramp Signla")
plt.xlabel("n")
plt.ylabel("r(n)")
plt.legend()
plt.show()
```

Exponential Sequence (Ab^n):

Definition: A discrete-time signal where each sample is multiplied by a constant factor b to get the next sample.

Formula: x[n]=Ab^n, where A is the amplitude and b is the base (growth/decay factor).

```python
A = 1
b = .5
t = np.arange(0, 10, .5)
y = A*np.exp(b*t)

plt.stem(t, y, label="Exponential Sequence")
plt.xlabel("Time(t)")
plt.ylabel("y(t)")
plt.legend()
plt.show()
```

Sine Sequence
Definition: Discrete-time versions of continuous-time sinusoidal signals. They are periodic (or almost periodic) and oscillate.
Sine: x[n]=Asin(ωn+ϕ)

```python
A = 1
x = np.arange(-10, 10, .25)
y = A*np.sin(x)

plt.plot(x, y, label="Sine Sequence")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

Cosine Sequence
Definition: Discrete-time versions of continuous-time sinusoidal signals. They are periodic (or almost periodic) and oscillate.
Cosine: x[n]=Acos(ωn+ϕ)

```python
A = 1
x = np.arange(-10, 11, .25)
y = A*np.cos(x)

plt.plot(x, y, label="Cosine Sequence")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
