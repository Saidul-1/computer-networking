 Demonstrates the effect of sampling, aliasing.

Sampling:
Definition: The process of converting a continuous-time signal x(t) into a discrete-time signal x[n] by taking its values at discrete, equally spaced points in time.

Process: x[n]=x(nT), where T is the sampling period (time between samples) and n is an integer index. The sampling frequency is F=1/T (samples per second or Hz).

Nyquist Theorem says: to accurately capture a signal without distortion, the sampling rate must be at least twice the maximum frequency present in the signal.

Aliasing:
Aliasing occurs when a signal is undersampled (fₛ ≤ 2fₕ), causing higher frequencies to appear as lower frequencies in the sampled signal.

Mitigation:

Anti-aliasing Filter (Low-Pass Filter): Before sampling, the analog signal is passed through an analog low-pass filter (LPF) to remove all frequency components above half the desired sampling frequency (F/2). This ensures that the Nyquist criterion is met for the filtered signal.
Increase Sampling Rate: If possible, sample at a much higher rate to avoid aliasing.

```python
import numpy as np
import matplotlib.pyplot as plt


#original signal
f_signal = 10
t = np.arange(0, 1, .001)
x_signal = np.sin(2*np.pi*f_signal*t)

#high sampling
f_high_sampling = 30
t_high_sampling = np.arange(0, 1, 1/f_high_sampling)
x_high_sampling = np.sin(2*np.pi*f_signal*t_high_sampling)

#low sampling
f_low_sampling = 15
t_low_sampling = np.arange(0, 1, 1/f_low_sampling)
x_low_sampling = np.sin(2*np.pi*f_signal*t_low_sampling)

#aliased signal
f_alias = f_signal-f_low_sampling
t_alias = t
x_alias = np.sin(2*np.pi*f_alias*t_alias)


# plt.figure(figsize=(12,8))
plt.subplot(3, 1, 1)
plt.plot(t, x_signal)
plt.stem(t_high_sampling, x_high_sampling)
plt.title("High Sampling")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()

plt.subplot(3, 1, 2)
plt.plot(t, x_signal)
plt.stem(t_low_sampling, x_low_sampling)
plt.title("Low Sampling")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()

plt.subplot(3, 1, 3)
plt.plot(t, x_signal, label="original signal")
plt.plot(t_alias, x_alias, label="aliased signal")
plt.stem(t_low_sampling, x_low_sampling)
plt.title("Aliasing")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```
