The impulse response of a discrete-time LTI system is h(n)={u(n)‚àíu(n‚àí5)}.
Determine the output of the system for the input x[n]=u(n), using the convolution
sum. 

Convolution is a mathematical operation that describes how an input signal ùë•[ùëõ] passes through a system with a given impulse response ‚Ñé[ùëõ] to produce the output ùë¶[ùëõ].
Discrete-Time LTI System:

LTI: Linear Time-Invariant.
Linearity: If input x1[n] gives output y1[n] and x2[n] gives y2[n], then ax1[n]+bx2[n] gives ay1[n]+by2[n].

Time-Invariance: If input x[n] gives output y[n], then x[n‚àík] gives y[n‚àík].

Impulse Response (h[n]): The output of an LTI system when the input is a discrete-time unit impulse Œ¥[n]. The impulse response completely characterizes an LTI system.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signals
n = np.arange(-2, 15)  # Time indices

# Unit step function
def u(n):
    return np.where(n >= 0, 1, 0)

# Impulse response
h_n = u(n) - u(n-5)

# Input signal
x_n = u(n)

# Convolution
y_n = np.convolve(x_n, h_n)
ny = np.arange(-4, len(n)*2-5)  # Adjusted time indices for output

# Plotting
plt.figure(figsize=(15, 5))

plt.subplot(3, 1, 1)
plt.stem(n, h_n)
plt.title('Impulse Response h[n] = u[n] - u[n-5]')
plt.xlabel('n')
plt.ylabel('h[n]')

plt.subplot(3, 1, 2)
plt.stem(n, x_n)
plt.title('Input Signal x[n] = u[n]')
plt.xlabel('n')
plt.ylabel('x[n]')

plt.subplot(3, 1, 3)
plt.stem(ny, y_n)
plt.title('Output Signal y[n] = x[n] * h[n]')
plt.xlabel('n')
plt.ylabel('y[n]')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def u(n):
    return np.where(n >= 0, 1, 0)

def convolve(x, h):
    len_y = len(x)+len(h)-1
    y = [0]*len_y

    for i in range(len_y):
        for k in range(len(x)):
            if(i-k >= 0 and i-k < len(h)):
                y[i] += x[k]*h[i-k]
    
    return y

n = np.arange(-4, 20)
x = u(n)
h = u(n)-u(n-5)

y = convolve(x, h)

plt.subplot(3, 1, 1)
plt.stem(n, x)

plt.subplot(3, 1, 2)
plt.stem(n, h)

ny = np.arange(-4, len(y)-4)  # Adjusted time indices for output

plt.subplot(3, 1, 3)
plt.stem(ny, y)
plt.show()
```
