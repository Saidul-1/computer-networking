Show that the highest rate of oscillation in a discrete-time sinusoidal is obtained
when ω=π. 

Rate of Oscillation:The rate of oscillation refers to how frequently the signal alternates between values — especially how rapidly it changes direction (peaks and zero crossings).
=> x[n] = cos(ωn+ϕ) = cos((ω + 2πk)n+ϕ) for any integer k
=> ω is the digital angular frequency, ranging from −π to π (or 0 to 2π). discrete-time signals are sampled, frequencies are periodic with period 2π. Frequencies outside this range are aliases of frequencies within this range. The Nyquist frequency (the maximum unique frequency) is: ω=π(radians/sample) This is the highest frequency that can be represented without aliasing.
=> In continuous-time, increasing Ω always leads to a higher frequency. In discrete-time, frequencies are periodic with 2π. This means ω and ω+2πk (for any integer k) represent the same discrete-time frequency. Due to this periodicity, the unique frequency range is typically taken as [−π,π] or [0,2π].Any frequency outside this range will appear as an alias within this range. For example, a continuous-time frequency sampled such that its discrete-time counterpart is 3π/2 will be indistinguishable from −π/2. A frequency of 3π will be indistinguishable from π.

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, 51, 1)

k = 0.1
omega1 = k*np.pi
omega2 = np.pi

x1 = np.cos(omega1*n)
x2 = np.cos(omega2*n)

plt.subplot(2, 1, 1)
plt.stem(n, x1)
plt.title("low oscillation at w=k*pi")
plt.xlabel("n")
plt.ylabel("x[n]")

plt.subplot(2, 1, 2)
plt.stem(n, x2)
plt.title("Highest oscillation at w=pi")
plt.xlabel("n")
plt.ylabel("x[n]")

plt.show()
```
